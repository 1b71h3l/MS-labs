## Challenge 01: Defective Items in a Manufacturing Batch
You work in a quality control department, and your team is responsible for inspecting a batch of 250 manufactured electronic devices. The batch is known to have a defect rate of 5%. Each device is inspected independently.

1. Identify what probability distribution is appropriet to this problem.
2. Using the `numpy.random` python library, simulate the inspection process for the batch of 250 devices.
3. Visualize the  probability density function of the number of defective devices. 



```python
# Your solution
import numpy as np
import matplotlib.pyplot as plt

rate = 0.05
devices = 250

defective_devices = np.random.binomial(devices, rate)

print("Number of defective devices:", defective_devices)

defective_devices = np.arange(0, devices + 1)
probabilities = np.random.binomial(devices, rate, size=devices + 1) / devices

plt.bar(defective_devices, probabilities)
plt.xlabel('Number of Defective Devices')
plt.ylabel('Probability')
plt.title('Probability Density Function')
plt.show()
```

### Part One: 
We want to model a chemical reaction system using an ODE model. The goal is to simulate the behavior of the reaction over time and estimate the reaction rates of different reactants.

Consider the following chemical reaction:

A + B â†’ C

Let $A_c$, $B_c$, and $C_c$ be the concentrations of A, B, and C, respectively, at time $t$. The rate of this chemical reaction can be modeled using the following ODE system:

\begin{cases} 
      \frac{d}{dt}{A_c}= -\alpha A_c B_c - \frac{1}{2} \beta C_c , \\
      \frac{d}{dt}{B_c}= -\alpha A_c B_c + \beta C_c, \\
      \frac{d}{dt}{C_c}= \frac{1}{2} \alpha A_c B_c,
\end{cases}



In this system, $alpha=0.1$ represents the rate constant for the reaction and $\beta=0.125$ is a modulator constant.


1. Model and Simulate this ODE system for a time duration of $t=20$ seconds using `scipy.integrate.odeint` and plot both the time evolution and phase plot of $A_c$, $B_c$, and $C_c$ for the initial condtions values: ${A_c}_0 = 1.0$ , ${B_c}_0 = 1.0$ , and ${C_c}_0 = 0.0$.
2. Experiment with another  initial concentrations of A (${A_c}_0 = 1.5$) and B (${B_c}_0 = 0.5$) to see how they affect the behavior of the reaction. Plot both the results of previous and current concentrations in one plot.    





```python
# Your solution
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
A0, B0 ,C0 = 1, 1, 0
# Everyone else, S0, is susceptible to infection initially.
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
alpha=0.1
beta=0.125
# A grid of time points (in days)
t = np.linspace(0, 20, 20)

# The SIR model differential equations.
def Chemie(y, t, N=1000, beta=0.125, alpha=0.1):
    #complete  this function to return the derivatives of each SIR compartment
    A, B , C = y[0], y[1], y[2]
    dAdt, dBdt , dCdt = -(alpha*B*A) -0.5*beta*C, (-(alpha*B*A))+(beta*C),(0.5*alpha*B*A)
    return [dAdt, dBdt, dCdt]

# Initial conditions vector
y0 = A0, B0 ,C0
# Integrate the SIR equations over the time grid, t.
#sol = call the odeint function to return your numerical solutions
Sol = odeint(Chemie, y0, t)

# Plot the data in the same plot on three separate curves for S(t), I(t) and R(t) 
#with labels 'Susceptible' for S(t), 'Infected' for I(t), and 'Recovered with immunity' for I(t)

A = Sol[:,0] 
B = Sol[:,1] 
C=Sol[:,2]

# Letâ€™s plot â€™rabbitâ€™ and â€™foxâ€™ populations as a function of time
plt.plot(t, A, label="A")
plt.plot(t, B, label="B")
plt.plot(t,C,label="C")
plt.xlabel("Time", fontsize=14)
plt.ylabel("Evolution", fontsize=14)
plt.legend();

    
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
A0, B0 ,C0 = 1, 1, 0  # old initial concentrations
A0_new, B0_new, C0_new = 1.5, 0.5, 0  # new initial concentrations
# Everyone else, S0, is susceptible to infection initially.
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
alpha = 0.1
beta = 0.125
# A grid of time points (in days)
t = np.linspace(0, 20, 20)

# The SIR model differential equations.
def Chemie(y, t, N=1000, beta=0.125, alpha=0.1):
    A, B , C = y[0], y[1], y[2]
    dAdt = -(alpha*B*A) - 0.5*beta*C
    dBdt = (-(alpha*B*A)) + (beta*C)
    dCdt = 0.5*alpha*B*A
    return [dAdt, dBdt, dCdt]

# Initial conditions vectors
y0 = A0, B0, C0
y0_new = A0_new, B0_new, C0_new

# Integrate the SIR equations over the time grid, t.
Sol = odeint(Chemie, y0, t)
Sol_new = odeint(Chemie, y0_new, t)

# Plot the data in the same plot on three separate curves for S(t), I(t) and R(t) 
# with labels 'Susceptible' for S(t), 'Infected' for I(t), and 'Recovered with immunity' for R(t)

A = Sol[:, 0] 
B = Sol[:, 1] 
C = Sol[:, 2]

A_new = Sol_new[:, 0] 
B_new = Sol_new[:, 1] 
C_new = Sol_new[:, 2]

# Plotting the results
plt.plot(t, A, label="A (Previous)")
plt.plot(t, B, label="B (Previous)")
plt.plot(t, C, label="C (Previous)")

plt.plot(t, A_new, linestyle="--", label="A (New)")
plt.plot(t, B_new, linestyle="--", label="B (New)")
plt.plot(t, C_new, linestyle="--", label="C (New)")

plt.xlabel("Time", fontsize=14)
plt.ylabel("Evolution", fontsize=14)
plt.legend()
plt.show()
```

### Part Two: 
We want to simulate the growth of an insect population that undergoes the following growth model:

$\frac{dP}{dt} = r* P * (1 - \frac{P}{K})$,

Where:

1. The initial population size at time $t=0$ is $P_0=100$.
2. The population growth rate $r=0.03$  is proportional to the current population size $P$.
 $K = 1000$ is the carrying capacity.

- Simulate this model in time periode of $t=120$ seconds.
- Analyze the population growth curve and observe how it evolves.
- What is the population size at $t=60$ seconds.

```python
#Your Solution
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from scipy.integrate import odeint 

P0=100
r=0.03
K=1000

sy.init_printing()

def InsectPopGrowth(x, time):
    return  (r*x*(1-(x/K)))

def LogExactSol(time_arr,IC):
    t = sy.symbols('t')
    f = sy.Function('P')
    # Solve the equation 
    diffEq = f(t).diff(t) -(r*f(t)*(1-(f(t)/K)))
    solve=sy.dsolve(diffEq, f(t), ics = {f(0): IC})
    #Get the exact numerical solution
    func = sy.lambdify(t, solve.rhs,'numpy') 
    yvals = func(time_arr)
    return yvals

#Get the numerical solution using odeint for initial Value IV
def NumericalSol(time_arr,IV):
    yvals = odeint(InsectPopGrowth, IV, time_arr)
    return yvals
 
ts = np.linspace(0.0,120.0, 60) # values of independent variable

#comppute exact and numerical solutions
yexact=LogExactSol(ts,P0)
ynum=NumericalSol(ts,P0)

#Plotting results
plt.scatter(ts,ynum, label='Numerical', color='r')
plt.plot(ts, yexact,label='Exact', color='b' )
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Logistic Population Growth')
plt.legend()
plt.show()


```
